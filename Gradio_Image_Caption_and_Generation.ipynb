```python
import gradio as gr
from transformers import pipeline
from diffusers import StableDiffusionPipeline
from PIL import Image
import io
import base64
import torch
```

```python
caption_pipeline = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base", device=0 if torch.cuda.is_available() else -1)
```

```python
def image_to_base64_str(pil_image):
    byte_arr = io.BytesIO()
    pil_image.save(byte_arr, format='PNG')
    byte_arr = byte_arr.getvalue()
    return str(base64.b64encode(byte_arr).decode('utf-8'))
```

```python
def captioner(image):
    result = caption_pipeline(image)[0]['generated_text']
    return result
```

```python
gr.close_all()
demo1 = gr.Interface(
    fn=captioner,
    inputs=[gr.Image(label="Upload image", type="pil")],
    outputs=[gr.Textbox(label="Caption")],
    title="Image Captioning with BLIP"
)
demo1.launch(share=True)
```

```python
gr.close_all()
```

```python
def generate(prompt):
    image = pipe(prompt, num_inference_steps=25).images[0]
    return image
```

```python
gr.close_all()
demo2 = gr.Interface(
    fn=generate,
    inputs=[gr.Textbox(label="Your prompt")],
    outputs=[gr.Image(label="Result")],
    title="Image Generation with Stable Diffusion"
)
demo2.launch(share=True)
```
