```python
import numpy as np

def positional_encoding(seq_length, embedding_dim):
    pos_enc = np.zeros((seq_length, embedding_dim))
    for pos in range(seq_length):
        for i in range(0, embedding_dim, 2):  # Even indices use sine, odd use cosine
            pos_enc[pos, i] = np.sin(pos / (10000 ** (i / embedding_dim)))
            if i + 1 < embedding_dim:
                pos_enc[pos, i + 1] = np.cos(pos / (10000 ** (i / embedding_dim)))
    return pos_enc

# Test the function
seq_length = 10  # Number of words in the sentence
embedding_dim = 8  # Embedding size per word
pos_encodings = positional_encoding(seq_length, embedding_dim)

print("Positional Encodings:\n", pos_encodings)
```

```python
import matplotlib.pyplot as plt

# Generate positional encodings
pos_encodings = positional_encoding(seq_length=50, embedding_dim=16)

# Plot the encodings
plt.figure(figsize=(10, 6))
plt.imshow(pos_encodings, cmap="viridis", aspect="auto")
plt.colorbar()
plt.xlabel("Embedding Dimension")
plt.ylabel("Word Position")
plt.title("Visualization of Positional Encoding")
plt.show()
```

```python
import numpy as np

def self_attention(query, key):
    """
    Compute attention scores using dot product similarity.
    """
    scores = np.dot(query, key.T)  # Dot product of query and key
    return scores

# Example sentence with two words (Let's, Go)
query = np.array([[1.2, 0.8], [0.5, 1.5]])  # Representing two words
key = np.array([[1.1, 0.9], [0.6, 1.3]])  # Same size as query

attention_scores = self_attention(query, key)
print("Self-Attention Scores:\n", attention_scores)
```
